# Hanoi Towers

```python
times = 0

def dfs(n):
    source = list(range(n, 0, -1))
    auxiliary = []
    target = []

    # Stack to simulate DFS
    stack = [(n, source, auxiliary, target, 'Start')]

    while stack:
        n, src, aux, tgt, path = stack.pop()

        # Base case: only one disk to move
        if n == 1:
            tgt.append(src.pop())
            print(f"{path} -> Move disk 1 from {src} to {tgt}")
            global times
            times = times + 1
            continue

        # Push the operations onto the stack in reverse order of execution
        # Move n-1 disks from auxiliary to target using source as auxiliary
        stack.append((n-1, aux, src, tgt, f"{path} -> Move disk {n} from {src} to {tgt}"))

        # Move the nth disk from source to target
        stack.append((1, src, aux, tgt, path))

        # Move n-1 disks from source to auxiliary using target as auxiliary
        stack.append((n-1, src, tgt, aux, path))



n = 3  # Number of disks
dfs(n)

print("Total number of moves: ", times)
```

```python
from collections import deque

times = 0

def bfs(n):
    initial_state = (list(range(n, 0, -1)), [], [])
    
    # Use a queue for BFS
    queue = deque([(initial_state, "Start")])
    visited = []
    visited.append(initial_state)
    
    # Helper function to generate the next states
    def get_next_states(state):
        pegs = [list(state[0]), list(state[1]), list(state[2])]
        next_states = []
        for i in range(3):
            if pegs[i]:
                disk = pegs[i].pop()
                for j in range(3):
                    if i != j and (not pegs[j] or pegs[j][-1] > disk):
                        new_pegs = [list(pegs[0]), list(pegs[1]), list(pegs[2])]
                        new_pegs[j].append(disk)
                        next_states.append((tuple(new_pegs[0]), tuple(new_pegs[1]), tuple(new_pegs[2]), (i, j, disk)))
                pegs[i].append(disk)
        return next_states
    
    # BFS loop
    while queue:
        state, path = queue.popleft()
        
        if len(state[2]) == n:
            print(path)
            return
        
        for next_state in get_next_states(state):
            new_state = (next_state[0], next_state[1], next_state[2])
            if new_state not in visited:
                visited.append(new_state)
                global times
                times = times + 1

                tower_source = '' 
                if next_state[3][0] + 1 == 1:
                    tower_source = 'A'
                
                if next_state[3][0] + 1 == 2:
                    tower_source = 'B'

                if next_state[3][0] + 1 == 3:
                    tower_source = 'C'

                tower_target = ''
                if next_state[3][1] + 1 == 1:
                    tower_target = 'A'
                
                if next_state[3][1] + 1 == 2:
                    tower_target = 'B'

                if next_state[3][1] + 1 == 3:
                    tower_target = 'C'

                queue.append((new_state, path + f" -> Mueve el disco {next_state[3][2]} de la torre {tower_source} a la torre {tower_target}\n"))
                


n = 6
bfs(n)

print(times)
```
