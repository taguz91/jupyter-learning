```python
# search
start = 'Ecuador'

path = []

def complete(node, path):
    path.append(node)

    for child in grafo[node]:
        if child not in path:
            complete(child, path)

complete(start, path)

print(path)
```

```python
# search
start = 'Ecuador'
end = 'Encebollado'

lifo = [{
    'node': start,
    'parent': '',
}]
visited = []

def search():
    while lifo:
        value = lifo.pop()
        node = value['node']

        if end == node:
            return value
        
        if node not in visited:
            visited.append(node)

            for child in grafo[node]:
                lifo.append({
                    'node': child,
                    'parent': value,
                })

result = search()
print(result)

```
