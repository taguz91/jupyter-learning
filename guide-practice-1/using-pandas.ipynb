# Pandas

```python
import csv
import pandas as pd

# open the report
with open("assets/poblacionMunicipios.csv", "r") as file:
    # read the report as dictionary
    reader = csv.DictReader(file)
    provinces = []

    # only get the province name to store in list
    for line in reader:
        provinces.append(line['Provincia'])
    
    # sort the list
    provinces.sort()

    postal_codes_report = []
    
    counter = 1
    # create the postal code for every province
    # with and auto incremental
    for province in provinces:
        postal_codes_report.append({
            'Provincia': province,
            'Codigo': province[:3] + '_' + str(counter).zfill(3)
        })

        counter += 1

    df = pd.DataFrame(postal_codes_report)
    print(df)

```

```python
provinces = [
    'Azuay',
    'Bolívar',
    'Cañar',
    'Carchi',
    'Chimborazo',
    'Cotopaxi',
    'El Oro',
    'Esmeraldas',
    'Galápagos',
    'Guayas',
    'Imbabura',
    'Loja',
    'Los Ríos',
    'Manabí',
    'Morona Santiago',
    'Napo',
    'Orellana',
    'Pastaza',
    'Pichincha',
    'Santa Elena',
    'Santo Domingo de los Tsáchilas',
    'Sucumbíos',
    'Tungurahua',
    'Zamora Chinchipe',
]

habitants = [
    801609,
    199078,
    227578,
    172828,
    471933,
    47021,
    714592,
    5539,
    28583,
    4391923,
    469879,
    485421,
    898652,
    1592840,
    192508,
    131675,
    182166,
    111915,
    3089473,
    385735,
    492969,
    199014,
    563532,
    110973,
]

# sum the total of habitants
total_habitants = sum(habitants)

percentage = []

# calculate the percentage in order
for province_habitants in habitants:
    percentage.append(province_habitants * 100 / total_habitants)
```

### Create the PIE

```python
import matplotlib.pyplot as plt

fig1, ax1 = plt.subplots()

ax1.pie(
    percentage,
    labels = provinces,
    autopct = '%1.1f%%',
    shadow = True,
    startangle = 90
)

# to ensure the circle
ax1.axis("equal")

plt.show()
```
