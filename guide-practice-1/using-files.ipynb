# Files

```python
# read the file
from io import open

# important use the latin-1 encoding 
# because the file text has accents
file = open("assets/FicheroALeer.txt", "r", encoding="latin-1")
text = file.read()
# print into the console the complete text
print(text)
```

```python
# return the cursor to start to reuse the file var
file.seek(0)
# read line by line
lines = file.readlines()

# count the lines and print 
print(f"El total de lineas es {len(lines)}")
```

```python
counter = 0

for line in lines:
    words = line.split(" ")
    for word in words:
        if word == "el":
            counter += 1

print(f"El artículo 'el' se repite {counter} veces")
```

```python
#  all text to lower
lower_text = text.lower()

# list of most common accents
accents = {
    'à': 'a',
    'á': 'a',
    'â': 'a',
    'ã': 'a',
    'ä': 'a',
    'å': 'a',
    'è': 'e',
    'é': 'e',
    'ê': 'e',
    'ë': 'e',
    'ì': 'i',
    'í': 'i',
    'î': 'i',
    'ï': 'i',
    'ò': 'o',
    'ó': 'o',
    'ô': 'o',
    'õ': 'o',
    'ö': 'o',
    'ù': 'u',
    'ú': 'u',
    'û': 'u',
    'ü': 'u',
}

# replace all accents
for key in accents.keys():
    sanitized_text = lower_text.replace(key, accents[key])

# write the file

file_target = open("assets/resultado.txt", "w")

file_target.writelines([
    f"Lineas: {len(lines)}\n",
    f"'el' se repite: {counter}\n",
    sanitized_text
])

# close the two files after execution
file.close()
file_target.close()
```
