```python
from sklearn.model_selection import train_test_split

xRedWineTrain, xRedWineTest, yRedWineTrain, yRedWineTest = train_test_split(
    dataRedWine,
    dataRedWine.index,
    random_state=4,
    # lo removemos para evitar el error
    # stratify=dataRedWine.index,
    train_size=0.6
)

print("Total data", dataRedWine.shape[0])
print("Total train data", xRedWineTrain.shape[0])
print("Total test data", xRedWineTest.shape[0])

# error
# The least populated class in y has only 1 member, which is too few. The minimum number of groups for any class cannot be less than 2.
```

```python
from sklearn.naive_bayes import MultinomialNB

naiveBayesRedWine = MultinomialNB()
naiveBayesRedWine.fit(xRedWineTrain, yRedWineTrain)

predictionsRedWine = naiveBayesRedWine.predict(xRedWineTest)
predictionsRedWine
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy:", accuracy_score(yRedWineTest, predictionsRedWine))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cmRedWine = confusion_matrix(yRedWineTest, predictionsRedWine)

print("Confusion Matrix:")
print(cmRedWine)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix = cmRedWine)
disp.plot()
```
