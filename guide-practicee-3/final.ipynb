```python
# con esto en mente agregamos una nueva categorizacion para manejar los datos

# 1 = mala calidad
# 2 = calidad media
# 3 = buena calidad
allWine['category_quality'] = allWine['quality'].apply(lambda x: 1 if x < 5 else 2 if x <= 7 else 3)
allWine.tail()

# revisamos la cantidad de vinos por calidad
countQuality = allWine["category_quality"].value_counts()
print(countQuality)
```

```python
allColumns = [
    "fixed acidity",
    "volatile acidity",
    "citric acid",
    "residual sugar",
    "chlorides",
    "free sulfur dioxide",
    "total sulfur dioxide",
    "density",
    "pH",
    "sulphates",
    "alcohol",
    "quality",
    "category",
    "category_quality"
]

# eliminamos los valores con calidad baja, para que nuestro algoritmo sea mas eficiente
# y prediga los vino de calidad media y alta
allWineWithoutQuality = pd.DataFrame(
    allWine[allColumns].loc[
        (allWine['category_quality'] > 1)
    ]
)

allWineWithoutQuality.describe()
```

```python
totalDeleted = len(allWine) - len(allWineWithoutQuality)
print("Se eliminaron: ", totalDeleted)
```

```python
# entrenamos el modelo con los datos que tenemos 

from sklearn.model_selection import train_test_split

allWineTrain = allWineWithoutQuality.set_index('category_quality')

xTrain, xTest, yTrain, yTest = train_test_split(
    allWineTrain,
    allWineTrain.index,
    random_state=1,
    # stratify=allWineWithoutQuality.index,
    train_size=0.7
)

print("Total data:", allWineWithoutQuality.shape[0])
print("Total training data:", xTrain.shape[0])
print("Total test data:", xTest.shape[0])
```

```python
# usamos naive bayes para el modelo
from sklearn.naive_bayes import MultinomialNB

naive_bayes = MultinomialNB()
naive_bayes.fit(xTrain, yTrain)

# test with predictions

predictions = naive_bayes.predict(xTest)
predictions
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy:", accuracy_score(yTest, predictions))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(yTest, predictions)

print("Confusion Matrix:")
print(cm)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```
